```python
import pandas as pd 
import random
import math
import numpy as np

data = pd.read_csv("iris.data",  names = ["A", "B", "C", "D", "E"])
data = data.sample(frac=1).reset_index(drop=True)
x = data.drop(columns=["E"])
y = data.drop(columns=["A", "B", "C", "D"])

y = y.replace({"Iris-setosa":1, "versicolor":2, "Iris-virginica":3})
cent = x.sample(3, replace=True)
```

```python
for m in range(0,500):
    udata1 = [];
    udata2 = [];
    udata3 = [];
    for i in range(0,149):
        a = math.sqrt(math.pow(float(cent.iloc[0,0] - x["A"][i]),2)+math.pow(float(cent.iloc[0,1]-x["B"][i]), 2)+math.pow(float(cent.iloc[0,2]-x["C"][i]), 2)+math.pow(float(cent.iloc[0,3]-x["D"][i]), 2))
        b = math.sqrt(math.pow(float(cent.iloc[1,0] - x["A"][i]),2)+math.pow(float(cent.iloc[1,1]-x["B"][i]), 2)+math.pow(float(cent.iloc[1,2]-x["C"][i]), 2)+math.pow(float(cent.iloc[1,3]-x["D"][i]), 2))
        c = math.sqrt(math.pow(float(cent.iloc[2,0] - x["A"][i]),2)+math.pow(float(cent.iloc[2,1]-x["B"][i]), 2)+math.pow(float(cent.iloc[2,2]-x["C"][i]), 2)+math.pow(float(cent.iloc[2,3]-x["D"][i]), 2))
        chk = [a,b,c]
        index = chk.index(min(chk))
        print(index)
        if index==0:
            udata1.append(x.iloc[i])
        elif index==1:
            udata2.append(x.iloc[i])
        elif index==2:
            udata3.append(x.iloc[i])
    x1,y1,z1,a1 = 0,0,0,0
    x2,y2,z2,a2 = 0,0,0,0
    x3,y3,z3,a3 = 0,0,0,0

    for i in range(0,len(udata1)):
        x1+=udata1[i][0]; y1+=udata1[i][1]; z1+=udata1[i][2]; a1+=udata1[i][3];

    for j in range(0,len(udata2)):
        x2+=udata2[j][0]; y2+=udata2[j][1]; z2+=udata2[j][2]; a2+=udata2[j][3];

    for k in range(0,len(udata3)):
        x3+=udata3[k][0]; y3+=udata3[k][1]; z3+=udata3[k][2]; a3+=udata3[k][3];

    cent.iloc[0,0]=x1/i; cent.iloc[0,1]=y1/i; cent.iloc[0,2]=z1/i; cent.iloc[0,3]=a1/i;
    cent.iloc[1,0]=x2/j; cent.iloc[1,1]=y2/j; cent.iloc[1,2]=z2/j; cent.iloc[1,3]=a2/j;
    cent.iloc[2,0]=x3/k; cent.iloc[2,1]=y3/k; cent.iloc[2,2]=z3/k; cent.iloc[2,3]=a3/k;
```
